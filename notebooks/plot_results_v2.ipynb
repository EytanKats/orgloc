```python
import pandas as pd
import seaborn as sns

from tabulate import tabulate
```

```python
df = pd.read_csv('/home/eytan/storage/staff/eytankats/projects/orgloc/experiments/202505141433_multilabel_basicunet_filteredaggmasksv2/results.csv')[:-2]
df['experiment'] = 'multilabel_basicunet_filteredaggmasksv2'
```

```python
def plot_results(data, columns, labels, title, aspect_ratio=1):

    df_filtered = data[data.columns.intersection(columns)]
    df_melted = pd.melt(df_filtered, "experiment")

    sns.set_theme(style="whitegrid")
    g = sns.catplot(kind='box', data=df_melted, x='variable', y='value', hue='experiment', height=6, aspect=aspect_ratio, legend=False, whis=(5, 95), width=.75)
    # g.set(ylim=(-70, 70))
    g.set(xlabel='', ylabel='detection offset error (mm)', title=title)
    g.set_xticklabels(labels, rotation=90)
```

```python
print(tabulate(df.tail(10), headers='keys', tablefmt='psql'))
```

```python
cols = ['left_mean', 'right_mean', 'superior_mean', 'posterior_mean', 'experiment']
lbls = ['left_mean', 'right_mean', 'superior_mean', 'inferior_mean']
plot_results(df, cols, lbls, "MEAN")
```
