```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('/home/kats/storage/staff/eytankats/projects/orgloc/experiments/202410282135_multilabel_basicunet_10000/test10k_margins/results.csv')[:-2]
df['experiment'] = '10000 training examples'
```

```python
def plot_results(data, columns, title, aspect_ratio=1):

    df_filtered = data[data.columns.intersection(columns)]
    df_melted = pd.melt(df_filtered, "experiment")

    sns.set(style="whitegrid")
    g = sns.catplot(kind='box', data=df_melted, x='variable', y='value', hue='experiment', height=6, aspect=aspect_ratio, legend=False, whis=(5, 95), width=.75)
    # g.set(ylim=(-70, 70))
    g.set(xlabel='', ylabel='detection offset error (mm)', title=title)
```

```python
cols = ['LF_mean', 'LB_mean', 'SF_mean', 'SB_mean', 'experiment']
plot_results(df, cols, "MEAN")
```
